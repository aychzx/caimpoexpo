```python
import requests
import pandas as pd
import os
```

```python

# Assuming the text file is named 'data.txt' and is located in the same directory as the notebook
# Adjust the file path and delimiter as needed
#the file path is always the same with the difference of the year and month 124 is the code for canada
# Function to generate file path dynamically
def generate_file_path(year, month):
    base_path = "C:\\Users\\Liamt\\Downloads\\C_M_H6_124_"
    return os.path.join(base_path + f"{year}{month:02d}.gz")

# Example usage
year = 2023
months = range(1, 13)  # Loop through all months from 1 to 12

# Read the file using the generated file path
# Initialize an empty list to store DataFrames
dfs = []

# Loop through each month, read the file, and append the DataFrame to the list
for month in months:
    file_path = generate_file_path(year, month)
    if os.path.exists(file_path):  # Check if the file exists
        try:
            # Read the file
            df_from_txt = pd.read_csv(file_path, delimiter='\s+', compression='gzip')  # Adjust the delimiter if necessary
            dfs.append(df_from_txt)
            
            # Print success message
            print(f"Successfully loaded file for {year}-{month:02d} into DataFrame.")
        
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
    else:
        print(f"File not found: {file_path}")  # Optional: Log missing files

# Concatenate all DataFrames into a single DataFrame
if dfs:  # Ensure there is at least one DataFrame in the list
    df_combined = pd.concat(dfs, ignore_index=True)
    print("All files successfully concatenated into a single DataFrame.")
    print(df_combined.head())
else:
    print("No files were loaded.")
```

```python
# so now that we have are final df I will create an easy to use sql database for my teamates
# they will be able to use sql queries to get the data they need put it in pandas df and graph it
# I will use sqlite3 to create the database
import sqlite3
```

```python
# Assuming df_final is your DataFrame in memory

# Step 1: Connect to (or create) the SQLite database
conn = sqlite3.connect('COMTRADE_data.db')  # This creates the database if it doesn't exist
cursor = conn.cursor()

# Step 2: Set SQLite PRAGMA settings for performance optimization
conn.execute("PRAGMA synchronous = OFF;")        # Reduce disk sync frequency for faster write
conn.execute("PRAGMA journal_mode = MEMORY;")     # Store journal in memory for faster transactions
conn.execute("PRAGMA cache_size = -64000;")       # Increase cache size (~64 MB here)
conn.execute("PRAGMA foreign_keys = OFF;")        # Temporarily disable foreign key constraints

# Step 3: Use chunked insertion to load data efficiently
# Chunk size
chunksize = 10000
total_rows = len(df_final)
total_chunks = (total_rows // chunksize) + 1  # Total number of chunks to process
progress_interval = 5  # 5% progress interval

# Calculate how many rows represent 5% of the total
rows_per_interval = (total_rows * progress_interval) // 100
next_progress = rows_per_interval  # Track the next milestone for printing

for i in range(0, len(df_final), chunksize):
    chunk = df_final.iloc[i:i+chunksize]
    chunk.to_sql('trade_data', conn, if_exists='append', index=False)
    print(f"Inserted chunk {i // chunksize + 1}")
    # Check if the current number of rows processed has reached the next milestone
    if i + chunksize >= next_progress:
        # Calculate the percentage completed and print progress
        progress_percentage = ((i + chunksize) / total_rows) * 100
        print(f"{progress_percentage:.1f}% complete")
        
        # Update next milestone
        next_progress += rows_per_interval

# Step 4: Reset PRAGMA settings to defaults after insertion is complete
conn.execute("PRAGMA foreign_keys = ON;")
conn.execute("PRAGMA synchronous = FULL;")
conn.execute("PRAGMA journal_mode = DELETE;")

# Step 5: Commit and close the connection
conn.commit()
conn.close()

print("Database created and data successfully loaded into the database.")
```
